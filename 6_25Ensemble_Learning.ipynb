# Ensemble Learning

강력한 하나의 모델보다, 다수의 약한 모델을 조합하면 더 정확한 예측이 가능

# 아래는 모두 Classification

## Voting Classifier

### Hard Voting Classifier

각 weak learner들의 예측 결과값을 가지고 다수결 투표

### Soft Voting Classifier

각 weak learner들의 예측 결과값을 평균냄

## Bagging (Bootstrap Aggregating)

bootstrap : random sampling을 통해 subset 생성

같은 종류의 classifier를 다양한 hyperparameter로 학습

### ******Random Forest***

Decision Tree를 classifier로 사용하는 Bagging 방식

여러 트리의 예측값을 평균내거나, 하나 고름

## Boosting

여러 classifier를 차례로 (sequentially) 연결해서 학습시킴

### AdaBoost (Adaptive Boost)

앞의 classifier가 예측하지 못했던 데이터의 가중치를 높임 (업데이트)
그 데이터를 다음 classifier에 넣음

### Gradient Boost

가중치 업데이트를 경사 하강법으로 함

XGBoost, LightGBM, CatBoost
