# NLP
Natural Language Processing

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import random
SEED = 42
random.seed(SEED)
np.random.seed(SEED)

import tensorflow as tf
import keras
keras.utils.set_random_seed(SEED)
```

## лҚ°мқҙн„° мӨҖл№„

```python
corpus = [
    'I am Student',
    'I love my dog',
    'I love my cat',
    'You love my dog',
    'Do you think my dog is amazing'
]
```

```python
# мҶҢл¬ёмһҗлЎң нҶөмқј

corpus = [sent.lower() for sent in corpus]
```

## Tokenize

```python
from tensorflow.keras.preprocessing.text import Tokenizer

tokenizer = Tokenizer(num_words=100, oov_token='<OOV>')   # oov : out of vocab
tokenizer.fit_on_texts(corpus)
```

```python
tokenizer.word_index
```

```python
tokenizer.index_word
```

## Sequenceнҷ”

```python
input_sequences = []
for sentence in corpus:
    token_list = tokenizer.texts_to_sequences([sentence])[0]
    for i in range(2, len(token_list) + 1):
        n_gram_sequence = token_list[:i]
        input_sequences.append(n_gram_sequence)
```

## Padding

```python
from keras.utils import pad_sequences

max_sequence_len = max(len(seq) for seq in input_sequences)
input_sequences = pad_sequences(
    input_sequences,
    maxlen=max_sequence_len,
    padding='pre'
)
```

## лҚ°мқҙн„° л¶„лҰ¬

```python
x = input_sequences[:, :-1]
y = input_sequences[:, -1]
```

## Encoding
one hot encoding

```python
from keras.utils import to_categorical
vocab_size = len(tokenizer.word_index) + 1   # padding 0 нҸ¬н•Ё
y = to_categorical(y, num_classes=vocab_size)
```

## лӘЁлҚё м •мқҳ

```python
from keras import layers

model = keras.Sequential([
    layers.Embedding(input_dim=vocab_size, output_dim=16),
    layers.LSTM(64),
    layers.Dense(32, activation='relu'),
    layers.Dense(vocab_size, activation='softmax')
])
```

### м»ҙнҢҢмқј

```python
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
model.summary()
```

## н•ҷмҠө

```python
history = model.fit(
    x,
    y,
    epochs=300,
    verbose=1
)
```

## Next Word Prediction

```python
def predict_next_word(seed_text):
    seed_text = seed_text.lower()
    token_list = tokenizer.texts_to_sequences([seed_text])[0]
    token_list = pad_sequences(
        [token_list],
        maxlen=max_sequence_len - 1,
        padding='pre'
    )
    predicted_prob = model.predict(token_list, verbose=0)
    predicted_index = np.argmax(predicted_prob)
    predicted_word = tokenizer.index_word.get(predicted_index, '<UNKNOWN>')
    return predicted_word
```

## н…ҢмҠӨнҠё

```python
test_sentences = [
    "i love my",
    "you love my",
    "do you think my dog is",
    "i am"
]

for text in test_sentences:
    next_word = predict_next_word(text)
    print(f"{text} -> {next_word}")
```
