# 외부 파일에서 데이터 읽어오기

```python
url = "https://raw.githubusercontent.com/devdio/flyai_datasets/refs/heads/main/titanic.csv"
df = pd.read_csv(url)
df.shape
```

```python
df.head()
```

```python
df[["Age", "Name"]]
```

```python
df.loc[(df["Age"] > 65) & (df["Sex"]=="male"), :]
```

##### 행 추가및 삭제

```python
# sibsp + parch = family 라는 행 추가
df["Family"] = df["SibSp"] + df["Parch"]
df[["Name", "Family", "SibSp", "Parch"]]
```

```python
# sibsp, parch 행 제거
df2 = df.drop(['SibSp', "Parch"], axis = 1)
df2
```

### 통계랑

```python
df["Age"].max(), df["Age"].min(), df["Age"].median()
```

### 그룹화 Groupby

```python
# 성별로 그룹을 나눈 후, 각각에서 Age의 중앙값 찾음
df.groupby("Sex")["Age"].median()
```

```python
# 성별로도, 좌석 클래스로도 그룹을 나눔
df.groupby(["Sex", "Pclass"])["Age"].mean()
```

### 결측치

```python
df["Age"].isna().sum()
# 비어있는 (true인) 갯수
```

```python
x = np.arange(10)
y = np.arange(10)
z = y^2
```

```python
plt.plot(x, y, 'r--')
plt.plot(x, z, 'g^')
plt.title("차트")
plt.xlabel('나이')
plt.ylabel('친구 수')
plt.show()
```

### subplot

```python
plt.figure(figsize = (6,3)) # inch 크기
plt.subplot(131)
plt.plot(x, x, "r--")
plt.subplot(132)
plt.plot(x, x^2, "g--")
plt.subplot(133)
plt.plot(x, x, "b--")
```

```python
x = np.linspace(0,2,100)  # 0과 2사이에서 100개 가져옴

fig, ax = plt.subplots(1,3, figsize=(6,3))
ax[0].plot(x,x, "r--")
ax[1].plot(x,x**2, "g--")
ax[2].plot(x,x, "b--")

```
