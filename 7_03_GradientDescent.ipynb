```python
import numpy as np

# f(x, y) = 2x^2 + y^2 + e^(4xy)
def f(x, y):
    return 2*x**2 + y**2 + np.exp(4*x*y)

# gradient
def grad_f(x, y):
    exp_term = np.exp(4*x*y)

    df_dx = 4*x + 4*y*exp_term
    df_dy = 2*y + 4*x*exp_term

    return df_dx, df_dy


# 초기값
x = 1.0
y = 1.0

# 설정
learning_rate = 0.005
num_iterations = 100

history = []

for i in range(num_iterations):
    df_dx, df_dy = grad_f(x, y)

    # gradient descent update
    x = x - learning_rate * df_dx
    y = y - learning_rate * df_dy

    current_f = f(x, y)
    history.append((i+1, x, y, current_f))

    print(f"iter {i+1:3d}: x = {x:.6f}, y = {y:.6f}, f(x,y) = {current_f:.6f}")


print("\nFinal result")
print(f"x = {x}")
print(f"y = {y}")
print(f"f(x,y) = {f(x, y)}")
```
