# Unlabeled Unsupervised Learning

## Clustering

unlabeled data에서 군집을 찾는것

### K-Means Clustering

K 개의 그룹으로 묶음


절차
1. K개의 centroid를 초기화
2. 각 데이터를 가장 가까운 centroid에 할당
3. 각 cluster에 속한 데이터들의 평균으로 centroid를 다시 계산
4. cluster 할당이 더 이상 변하지 않거나 변화가 매우 작아질 때까지 반복

SSE (Sum of Square Errors) 를 측정해서, 감소폭이 급격히 줄어드는 지점을 최적의 K로 선택함. (Elbow Method)


## Dimensionality Reduction

정보 손상 최소화로 양을 줄이는 방법

Dimension이 증가하면 정보의 밀도가 감소하므로, overfitting이 생길수 있다.

### PCA (Principal Component Analysis)

데이터가 가장 넓게 퍼져있는 방향 (즉 분산이 큰 방향)을 찾아서 새로운 축(주성분 = principal component)으로 잡음.

첫번째 축에 직교하고 남은 분산을 최대한 보존하는 두번째 축을 찾음




PCA는 차원 축소, 시각화, 노이즈 감소, 학습 속도 향상에 사용. 단, feature scale의 영향을 많이 받기 때문에 보통 standardization 후 적용.

### **t**-SNE

고차원 데이터를 2-3차원으로 줄임

고차원 데이터 시각화
